```python
def make_correl(ryd_pop_per_param, rearrangement_mask_per_param, connected_corr=True, show_plots = False):
    """ Use Rydberg populations and rearrangement mask as inputs to return site-averaged 2D correlation function"""
    
    num_params = ryd_pop_per_param.shape[2]
    
    # Average pair correlation for each sweep parameter postselecting on defect number using rearrangement_mask
    n_ij_kl_corr_per_param = np.zeros((ryd_pop_per_param.shape[0], ryd_pop_per_param.shape[1], ryd_pop_per_param.shape[0], ryd_pop_per_param.shape[1], ryd_pop_per_param.shape[2]))

    # Average single site occupation for each sweep parameter postselecting on defect number using rearrangement_mask
    n_ij_per_param = np.zeros((ryd_pop_per_param.shape[0], ryd_pop_per_param.shape[1], ryd_pop_per_param.shape[2]))
    
    reps_included = np.sum(rearrangement_mask_per_param, axis=-1)
    
    # Construct correlations for each parameter postselecting on repetitions
    for m in range(num_params):
        n_ij_kl_corr_per_param[:,:,:,:,m] = np.einsum("ijn,kln", ryd_pop_per_param[:, :, m, rearrangement_mask_per_param[m,:]], ryd_pop_per_param[:, :, m, rearrangement_mask_per_param[m,:]]) / reps_included[m]
        n_ij_per_param[:,:,m] = np.mean(ryd_pop_per_param[:, :, m, rearrangement_mask_per_param[m,:]], axis = -1) 

    n_ij_kl_avg_per_param = np.diagonal(np.einsum("ijk,lmn", n_ij_per_param, n_ij_per_param), axis1=2, axis2=5)
    g2_ij_kl_per_param = n_ij_kl_corr_per_param - n_ij_kl_avg_per_param

    # Average g2_ij_kl over all pairs separated by the same horizontal and vertical distance
    avg_corr_per_param = np.zeros((2 * ryd_pop_per_param.shape[0] - 1,
               2 * ryd_pop_per_param.shape[1] - 1, num_params))
    
    num_contrib_per_param = np.zeros(avg_corr_per_param.shape)

    for m in range(num_params):
        for i in range(ryd_pop_per_param.shape[0]):
            for j in range(ryd_pop_per_param.shape[1]):
                for k in range(ryd_pop_per_param.shape[0]):
                    for l in range(ryd_pop_per_param.shape[1]):
                        dist_col = (k - i) + ryd_pop_per_param.shape[0] - 1
                        dist_row = (l - j) + ryd_pop_per_param.shape[1] - 1
                
                        # Keep track of number of sites separated by a particular horizontal / vertical distance
                        num_contrib_per_param[dist_col, dist_row, m] += 1 
                        if connected_corr:
                            avg_corr_per_param[dist_col, dist_row, m] += g2_ij_kl_per_param[i, j, k, l, m]
                        else:
                            avg_corr_per_param[dist_col, dist_row, m] += n_ij_kl_corr_per_param[i, j, k, l, m]

    contrib_mask = num_contrib_per_param > 0
    avg_corr_per_param[contrib_mask] /= num_contrib_per_param[contrib_mask] # Normalize by number of pairs that contribute
    
    if show_plots:
        plt.figure(figsize=[120, 40])
        for p in range(num_params):
            plt.subplot(1,num_params,p+1)
            if connected_corr:
                v_min, v_max = -0.25, 0.25
            else:
                v_min, v_max = 0, 0.5
            
            plt.imshow(avg_corr_per_param[:, :, p].T, origin='lower', vmin=v_min, vmax=v_max, cmap='seismic')
#            plt.colorbar()
    
    return avg_corr_per_param
```

```python
avg_corr_per_param = make_correl(load_file['rydberg_data'], rearrangement_mask_per_param, show_plots=True)
```

```python
def h_correl(avg_corr_per_param, show_plots = False):
    '''Correlation length per sweep parameter as function of horizontal distance'''
    
    rectify_plot = True
    
    num_params = avg_corr_per_param.shape[-1]
    num_h_sites = avg_corr_per_param.shape[1] - avg_corr_per_param.shape[1]//2
    horiz_corr_per_param = np.zeros((num_h_sites, num_params))
    
    cross_section = int((avg_corr_per_param.shape[0]-1)/2)
    for i in range(num_h_sites):
        horiz_corr_per_param[i, :] = avg_corr_per_param[cross_section, avg_corr_per_param.shape[1]//2+i, :]
    
    h_corr_lengths = []

    rectify_vector = np.ones(num_h_sites)
    if rectify_plot:
        rectify_vector[1::2]=-1

    if show_plots:
        plt.figure('Horizontal Correlation Length')
        plt.clf()
        plt.figure(figsize=[10, 7])
        
    for p in range(num_params):
        try:
            popt, pcov = curve_fit(exp, np.arange(1, num_h_sites), np.abs(horiz_corr_per_param[1:, p]), p0=(0.1, 2))
        except: # If correlations fall off too quickly for fit, just make them zero
            popt = np.array([0,0])
            pcov = np.array([0,0])
        
        if show_plots:
            plt.semilogy(range(num_h_sites), rectify_vector*horiz_corr_per_param[:, p], marker='o', ls='', c=color_map[p])
            plt.semilogy(np.arange(num_h_sites), exp(np.arange(num_h_sites), *popt), c=color_map[p], ls='dotted', label='$\\xi = %.2f$ sites' %popt[1])
        h_corr_lengths.append(popt[1])

    if show_plots:
        plt.legend(bbox_to_anchor=(1,1), loc='upper right')
        plt.ylim(1e-2, 0.5)
        plt.xlabel("Horizontal Distance")
        plt.ylabel("Rydberg-Rydberg Correlation")
        plt.title('Horizontal Correlation Length')
    
    return np.array(h_corr_lengths)
```

```python
h_corr_lengths = h_correl(avg_corr_per_param, show_plots=True)
```

```python
def v_correl(avg_corr_per_param, show_plots=False):
    '''Correlation length per sweep parameter as function of vertical distance'''
    
    rectify_plot = True
    
    num_params = avg_corr_per_param.shape[-1]
    num_v_sites = avg_corr_per_param.shape[0] - avg_corr_per_param.shape[0]//2
    vert_corr_per_param = np.zeros((num_v_sites, num_params))

    cross_section = int((avg_corr_per_param.shape[1]-1)/2)
    for i in range(num_v_sites):
        vert_corr_per_param[i, :] = avg_corr_per_param[avg_corr_per_param.shape[0]//2+i, cross_section, :]

    v_corr_lengths = []

    rectify_vector = np.ones(num_v_sites)
    if rectify_plot:
        rectify_vector[1::2]=-1

    if show_plots:
        plt.figure('Vertical Correlation Length')
        plt.clf()
        plt.figure(figsize=[10, 7])
    
    for p in range(num_params):
        try:
            popt, pcov = curve_fit(exp, np.arange(1, num_v_sites), np.abs(vert_corr_per_param[1:, p]), p0=(0.1, 2))
        except: # If correlations fall off too quickly for fit, just make them zero
            popt = np.array([0,0])
            pcov = np.array([0,0])
            
        if show_plots:
            plt.semilogy(range(num_v_sites), rectify_vector*vert_corr_per_param[:, p], marker='o', ls='', c=color_map[p])    
            plt.semilogy(np.arange(num_v_sites), exp(np.arange(num_v_sites), *popt), c=color_map[p], ls='dotted', label='$\\xi = %.2f$ sites' %popt[1])
    
        v_corr_lengths.append(popt[1])
    
    if show_plots:
        plt.legend(bbox_to_anchor=(1,1), loc='upper right')
        plt.ylim(1e-4, 0.5)
        plt.xlabel("Vertical Distance")
        plt.ylabel("Rydberg-Rydberg Correlation")
        
    return np.array(v_corr_lengths)
```

```python
v_corr_lengths = v_correl(avg_corr_per_param, show_plots=True)
```

```python
def coarse_grain(ryd_pop_per_param):
    '''Coarse grain to construct Ising order parameter'''

    ryd_pad = np.pad(ryd_pop_per_param,((1,1),(1,1), (0,0),(0,0)), 'constant', constant_values= -1)

    ising_order = np.zeros(ryd_pop_per_param.shape)

    for i in range(ryd_pop_per_param.shape[0]):
        for j in range(ryd_pop_per_param.shape[1]):
            num_contrib = 0
            if np.mean(ryd_pad[i-1+1,j+1,:,:], axis=(-1,-2)) > -0.5:
                ising_order[i,j,:,:] += (-1)**(i+j)*(ryd_pad[i+1,j+1,:,:]-ryd_pad[i-1+1,j+1,:,:])
                num_contrib += 1
            if np.mean(ryd_pad[i+1+1,j+1,:,:], axis=(-1,-2)) > -0.5:
                ising_order[i,j,:,:] += (-1)**(i+j)*(ryd_pad[i+1,j+1,:,:]-ryd_pad[i+1+1,j+1,:,:])
                num_contrib += 1
            if np.mean(ryd_pad[i+1,j-1+1,:,:], axis=(-1,-2)) > -0.5:
                ising_order[i,j,:,:] += (-1)**(i+j)*(ryd_pad[i+1,j+1,:,:]-ryd_pad[i+1,j-1+1,:,:])
                num_contrib += 1
            if np.mean(ryd_pad[i+1,j+1+1,:,:], axis=(-1,-2)) > -0.5:
                ising_order[i,j,:,:] += (-1)**(i+j)*(ryd_pad[i+1,j+1,:,:]-ryd_pad[i+1,j+1+1,:,:])
                num_contrib += 1
            
            ising_order[i,j,:,:] /= num_contrib
            
    return ising_order
```

```python
# Correlations in radially averaged distance units
def make_radial_correl(avg_corr):

    center_col = (avg_corr.shape[0]-1)//2
    center_row = (avg_corr.shape[1]-1)//2

    radial_distance = []
    radial_values = []
    radial_contrib = []

    for i in range(avg_corr.shape[0]):
        for j in range(avg_corr.shape[1]):
            distance = round(np.sqrt(np.square(i-center_col)+np.square(j-center_row)),5)
            distance_index = np.where(radial_distance==distance)[0]
            if len(distance_index) == 0:
                radial_distance.append(distance)
#                 radial_values.append(np.abs(avg_corr[i,j]))
                radial_values.append(avg_corr[i,j])
                radial_contrib.append(1)
            else:
                distance_index = distance_index[0]
                radial_values[distance_index]+=avg_corr[i,j]
                radial_contrib[distance_index]+=1
    for n in range(len(radial_distance)):
        if radial_contrib[n] > 0:
            radial_values[n] /= radial_contrib[n]
    return np.array(radial_distance), np.array(radial_values)
```

```python
def ir_correl(ising_corr_per_param, show_plots=False):
    '''Ising correlation length per sweep parameter as function of radial distance'''
    
    num_params = ising_corr_per_param.shape[2]

    for m in range(num_params):
        if m == 0:
            ising_radial_distance, ising_radial_val = make_radial_correl(ising_corr_per_param[:,:,m])
            ising_radial_values = np.zeros((len(ising_radial_val), num_params))
            ising_radial_values[:,m] = ising_radial_val
        else:
            ising_radial_distance, ising_radial_values[:,m] = make_radial_correl(ising_corr_per_param[:,:,m])

    ising_corr_lengths = []
    
    if show_plots:
        plt.figure('Ising Correlation Length')
        plt.clf()
        plt.figure(figsize=[10, 7])

    for p in range(num_params):
        try:
            popt, pcov = curve_fit(exp, ising_radial_distance[:-1], ising_radial_values[:-1,p], p0=(0.1, 2))
        except: # If correlations fall off too quickly for fit, just make them zero
            popt = np.array([0,0])
            pcov = np.array([0,0])
            
        if show_plots:
            plt.semilogy(ising_radial_distance[:-1], ising_radial_values[:-1,p], marker='o', ls='', c=color_map[p])
            plt.semilogy(ising_radial_distance[:-1], exp(ising_radial_distance[:-1], *popt), c=color_map[p], ls='dotted', 
                         label='$\\xi = %.2f$ sites' %popt[1])
    
        ising_corr_lengths.append(popt[1])

    if show_plots:
        plt.legend(bbox_to_anchor=(1,1), loc='upper right')
        plt.ylim(1e-3, 1)
        plt.xlim(0,15)
        plt.xlabel("Radial Distance")
        plt.ylabel("Ising Correlation")
    
    return np.array(ising_corr_lengths)
```

# Collate correlation lengths and analyze universal scaling

```python
load_sweeps = np.array([15, 21, 30, 42, 60, 85, 120])

# Slowest sweeps
load_file = np.load('KZ_data_16x16_%d_MHz_per_us.npz' %load_sweeps[0])
print('Loading sweep 1 of %d: %d MHz/us' %(len(load_sweeps), load_sweeps[0]))

params_slow = load_file['params']
ryd_pop_per_param = np.array(load_file['rydberg_data'], dtype=int)
rearrangement_mask_per_param = np.ones(load_file['rearrange_mask'].shape, dtype=bool)
avg_corr_per_param = make_correl(ryd_pop_per_param, rearrangement_mask_per_param, connected_corr=True, show_plots=False)
h_corr_slow = h_correl(avg_corr_per_param, show_plots=False)
v_corr_slow = v_correl(avg_corr_per_param, show_plots=False)
    
ising_order = coarse_grain(ryd_pop_per_param)
ising_corr_per_param = make_correl(ising_order, rearrangement_mask_per_param, connected_corr=True, show_plots=False)
ising_corr_slow = ir_correl(ising_corr_per_param, show_plots=False)
ising_h_corr_slow = h_correl(ising_corr_per_param, show_plots=False)
ising_v_corr_slow = v_correl(ising_corr_per_param, show_plots=False)

# Rest of the faster sweeps
for n in np.arange(len(load_sweeps)-1)+1:
    load_file = np.load('KZ_data_16x16_%d_MHz_per_us.npz' %load_sweeps[n])
    print('Loading sweep %d of %d: %d MHz/us' %(n+1, len(load_sweeps), load_sweeps[n]))
    
    ryd_pop_per_param = np.array(load_file['rydberg_data'], dtype=int)
    rearrangement_mask_per_param = np.ones(load_file['rearrange_mask'].shape, dtype=bool)
    avg_corr_per_param = make_correl(ryd_pop_per_param, rearrangement_mask_per_param, connected_corr=True, show_plots=False)
    h_corr_lengths = h_correl(avg_corr_per_param, show_plots=False)
    v_corr_lengths = v_correl(avg_corr_per_param, show_plots=False)
    
    ising_order = coarse_grain(ryd_pop_per_param)
    ising_corr_per_param = make_correl(ising_order, rearrangement_mask_per_param, connected_corr=True, show_plots=False)
    ising_corr_lengths = ir_correl(ising_corr_per_param, show_plots=False)
    ising_h_corr_lengths = h_correl(ising_corr_per_param, show_plots=False)
    ising_v_corr_lengths = v_correl(ising_corr_per_param, show_plots=False)
    
    if n == 1:
        params = load_file['params']
        h_corr_values = np.zeros((len(h_corr_lengths),len(params)-1))
        v_corr_values = np.zeros((len(v_corr_lengths),len(params)-1))
        ising_corr_values = np.zeros((len(ising_corr_lengths),len(params)-1))
        ising_h_corr_values = np.zeros((len(ising_h_corr_lengths),len(params)-1))
        ising_v_corr_values = np.zeros((len(ising_v_corr_lengths),len(params)-1))
        
    h_corr_values[:,n-1] = h_corr_lengths
    v_corr_values[:,n-1] = v_corr_lengths
    ising_corr_values[:,n-1] = ising_corr_lengths
    ising_h_corr_values[:,n-1] = ising_h_corr_lengths
    ising_v_corr_values[:,n-1] = ising_v_corr_lengths    
```
